```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
lbf_stats_hist.head()
```

```python
lbf_stats_hist_u=lbf_stats_hist["torneo"].unique()
print(lbf_stats_hist_u)
```

```python
torneo_ = {'Apertura':1, 'Clausura':2}
lbf_stats_hist["torneo"]=lbf_stats_hist["torneo"].map(torneo_)
```

```python
lbf_stats_hist.head()
```

```python
teams=lbf_stats_hist["club"].unique()
print(lbf_stats_hist_u)
```

```python
len(teams)
```

```python
team_mapping = {team: idx + 1 for idx, team in enumerate(teams)}
```

```python
team_mapping
```

```python
lbf_stats_hist["club"]=lbf_stats_hist["club"].map(team_mapping)
```

```python
lbf_stats_hist.head()
```

```python
seasons = lbf_stats_hist["temporada"].unique()
```

```python
season_mapping = {season: idx +1 for idx, season in enumerate(seasons) }
```

```python
print(season_mapping)
```

```python
lbf_stats_hist["temporada"] = lbf_stats_hist["temporada"].map(season_mapping)
```
