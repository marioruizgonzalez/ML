```python
# YouTube Libreto y C√≥digo para NURI - Estilo CRM Joyer√≠a

"""
üé• T√≠tulo del video: ¬øD√≥nde conviene invertir en Airbnb en CDMX? (Con datos reales)
Duraci√≥n sugerida: 8-10 minutos
Estilo: Profesional, educativo, sobrio y detallado.
Formato: Narraci√≥n + Jupyter Notebook (pantalla compartida)
Objetivo: Ense√±ar c√≥mo tomar decisiones basadas en datos reales cruzando precio y demanda.
"""

# ---
# üé¨ Introducci√≥n (Narraci√≥n en c√°mara)
"Hoy vamos a analizar datos reales de Airbnb en Ciudad de M√©xico. Queremos saber: ¬øEn qu√© zonas conviene m√°s tener una propiedad en renta? Esta informaci√≥n puede ayudarte a tomar decisiones m√°s inteligentes, tanto si inviertes como si solo est√°s aprendiendo ciencia de datos."

# Este an√°lisis tambi√©n es un excelente ejemplo de c√≥mo aplicar estad√≠stica descriptiva y pensamiento cr√≠tico usando datos reales.

# ---
# üßæ 1. Importar librer√≠as y cargar los datos
"Empezamos con lo esencial. Vamos a importar pandas para manejo de datos, matplotlib y seaborn para visualizaci√≥n. Tambi√©n cargamos nuestro dataset."

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Estilo visual
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Cargar el dataset
file_path = r"C:\Users\ShePlay\Downloads\archive\listings.csv"  # Aseg√∫rate de estar en el mismo directorio

df = pd.read_csv(file_path)
df.head()
```

```python
# Nugget: "En ciencia de datos, la visualizaci√≥n es m√°s que est√©tica, es una herramienta para pensar mejor."

# ---
# üßπ 2. Limpieza de datos
"Lo primero que hago es limpiar el dataset. Vamos a eliminar columnas que no usamos y filtrar registros con precios extremos o sin rese√±as."

# Eliminar columna vac√≠a
df.drop(columns=['neighbourhood_group'], inplace=True)

# Filtrar propiedades sin rese√±as
df = df[df['number_of_reviews'] > 0]

# Convertir tipo de dato de precio (si estuviera como string con s√≠mbolos)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Filtrar precios extremos para evitar distorsi√≥n
df = df[df['price'] < 5000]
```

```python

# ---
# üìä 3. Agrupar por delegaci√≥n y tipo de propiedad
"Agrupamos por delegaci√≥n y por tipo de hospedaje. Queremos entender el comportamiento de casas, departamentos, cuartos privados o habitaciones de hotel en cada zona."

zona_stats = df.groupby(['neighbourhood', 'room_type']).agg({
    'id': 'count',
    'number_of_reviews': 'sum',
    'reviews_per_month': 'mean',
    'price': 'mean'
}).reset_index().rename(columns={'id': 'num_props'})

zona_stats.head()

# Matem√°tica detr√°s: estamos usando estad√≠sticas descriptivas b√°sicas ‚Äî suma, media, conteo ‚Äî para generar una m√©trica de comportamiento por zona.

# ---
# üìà 4. Crear m√©trica de "potencial Airbnb"
"Ahora creamos una m√©trica que llamamos 'potencial': es una estimaci√≥n simple multiplicando la cantidad de propiedades por sus reviews por mes."

zona_stats['potencial'] = zona_stats['num_props'] * zona_stats['reviews_per_month']

# Esto es una proxy de flujo de actividad. Si hay muchas propiedades y muchas rese√±as por mes, hay alta rotaci√≥n y, por tanto, oportunidad.

zona_stats.sort_values('potencial', ascending=False).head(10)

# Nugget: "Las rese√±as por mes nos dan una idea del volumen de hu√©spedes reciente. Es un buen term√≥metro de rotaci√≥n."

# ---
# üîç 5. Visualizaci√≥n general
"Vamos a visualizar c√≥mo se comportan las delegaciones m√°s activas por tipo de hospedaje."

sns.barplot(data=zona_stats.sort_values('potencial', ascending=False).head(20), 
            x='potencial', y='neighbourhood', hue='room_type')
plt.title('Top zonas de CDMX por potencial Airbnb')
plt.xlabel('Potencial = propiedades √ó reviews/mes')
plt.ylabel('Delegaci√≥n')
plt.legend(title='Tipo de hospedaje')
plt.tight_layout()
plt.show()
```

```python
# üìà 6. Tendencia con regresi√≥n
"Para cerrar, quiero ver si existe una tendencia entre n√∫mero de propiedades y reviews mensuales. Usamos una regresi√≥n lineal para visualizar la correlaci√≥n."

sns.lmplot(data=zona_stats, x='num_props', y='reviews_per_month', hue='room_type', aspect=1.5)
plt.title('Relaci√≥n entre n√∫mero de propiedades y reviews por mes')
plt.xlabel('N√∫mero de propiedades')
plt.ylabel('Reviews mensuales promedio')
plt.tight_layout()
plt.show()
```
