```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]
y3 = [15, 5, 21, 15, 15]
```

```python
plt.plot(x, y1, label="Line 1", color="blue", linestyle='--', marker='o', markersize=8, linewidth=2)
plt.plot(x, y2, label="Line 2", color="green", linestyle='--', marker='o', markersize=8, linewidth=2)
plt.plot(x, y3, label="Line 2", color="red", linestyle='--', marker='o', markersize=8, linewidth=2)
plt.title("Example One")
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title("Multiple Lines in a Single Plot")
plt.legend()

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]
y3 = [15, 5, 21, 15, 15]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 4))

axes[0].plot(x, y1, color="blue", linestyle='--', marker='o', markersize=8, linewidth=2)
axes[0].set_title("First Plot")

axes[1].plot(x, y2, color="green", linestyle='--', marker='o', markersize=8, linewidth=2)
axes[1].set_title("Second Plot")

axes[2].plot(x, y3, color="red", linestyle='--', marker='o', markersize=8, linewidth=2)
axes[2].set_title("Third Plot")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')

plt.title("3D Surface Plot")

plt.show()
```

```python
# Create a DataFrame
data = {'X': [1, 2, 3, 4, 5], 'Y': [10, 20, 25, 30, 40]}
df = pd.DataFrame(data)

# Plot
df.plot(x='X', y='Y', kind='line', legend=False)

plt.title("Simple Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
# Data
data = {'X': [1, 2, 3, 4, 5], 'Y': [10, 20, 25, 30, 40]}
df = pd.DataFrame(data)

# Plot with customization
df.plot(x='X', y='Y', kind='line', color='red', linestyle='--', marker='o', markersize=8, linewidth=2, legend=False)

plt.title("Customized Line Plot")
plt.show()
```

```python
# Data
data = {'X': [1, 2, 3, 4, 5], 'Y': [10, 20, 25, 30, 40]}
df = pd.DataFrame(data)

# Plot
ax = df.plot(x='X', y='Y', kind='line', label="Data Line")

# Add labels and title
ax.set_title("Sample Plot")
ax.set_xlabel("X-Axis Label")
ax.set_ylabel("Y-Axis Label")
ax.legend()

plt.show()
```
